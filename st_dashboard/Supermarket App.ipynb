```python
import pandas as pd
import matplotlib.pyplot as ply
import numpy as np
import seaborn as sns

sns.set()
%matplotlib inline
```

```python
supermarket_df = pd.read_csv('supermarket.csv')
supermarket_df
```

```python
supermarket_df.info()
```

```python
#changing store_id to data type str
supermarket_df.store_id = supermarket_df.store_id.astype(str)
supermarket_df.store_area = supermarket_df.store_area.astype(str)
```

```python
supermarket_df.info()
```

```python
supermarket_df.describe()
```

```python
sns.boxplot(data=supermarket_df, x='items_available');
```

```python
sns.boxplot(data=supermarket_df, x='daily_customer_count');
```

```python
sns.boxplot(data=supermarket_df, x='store_sales');
```

```python
#Top Stores
data = supermarket_df.sort_values(by='store_sales', ascending=False)
data
```

```python
sns.scatterplot(data=data,x="items_available", y="store_sales");
```

```python
sns.scatterplot(data=data,x="daily_customer_count", y="store_sales");
```
